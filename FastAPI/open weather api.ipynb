```python
import requests
url = 'https://api.openweathermap.org/data/2.5/weather'
#parameter untuk API (misalnya nama kota, API key)
params = {'q': 'Kota Tangerang',
          'appid': '369086975a14e660eb7a4cf7742d7b75', #API key
          'units': 'metric'}
#membuat permintaan GET ke API
response = requests.get(url, params=params)
#cek apakah permintaan berhasil
if response.status_code == 200:
    data = response.json()
    city = data['name']
    temperature = data['main']['temp']
    description = data['weather'][0]['description']
    print(f'Cuaca di {city}:')
    print(f"Suhu: {temperature}°C")
    print(f"Deskripsi: {description}")
else:
    print('Terjadi kesalahan saat mengakses API.')
          
```

```python
import requests

# Fungsi untuk mendapatkan daftar simbol mata uang yang didukung
def get_supported_symbols():
    url = 'http://api.exchangeratesapi.io/v1/symbols'
    params = {
        'access_key': '21fc7e241875781b8b8ff8577d97607d'  # Masukkan API key Anda di sini
    }
    
    # Lakukan request GET ke API
    response = requests.get(url, params=params)
    
    # Periksa status code dan kembalikan data jika sukses
    if response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Request failed with status code {response.status_code}"}

# Panggil fungsi dan cetak daftar simbol mata uang
symbols_result = get_supported_symbols()
# Periksa apakah ada error
if 'error' not in symbols_result:
    # Dapatkan daftar simbol
    symbols = symbols_result.get('symbols', {})
    
    # Cetak simbol mata uang per baris
    for symbol, description in symbols.items():
        print(f"{symbol}: {description}")
else:
    print(symbols_result['error'])
```
